```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

dataset import

```python
url = 'https://raw.githubusercontent.com/galudSla/stroke-prediction/main/healthcare-dataset-stroke-data.csv'
df = pd.read_csv(url)
print (df.head())
```

dataset general information

```python
df.info()
print(df.isnull().sum())
```

dataset numerical data distribution

```python
for attr in df.columns:
    if attr =='stroke' or attr == 'id' or attr == 'gender' or attr == 'ever_married' or attr =='work_type' or attr =='Residence_type' or attr =='smoking_status':
        continue
    print(df[attr].describe())
    plt.figure(figsize=(9, 8))
    sns.distplot(df[attr], color='g', bins=100, hist_kws={'alpha': 0.4});
```

dataset numerical data histogram

```python
list(set(df.dtypes.tolist()))
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```
